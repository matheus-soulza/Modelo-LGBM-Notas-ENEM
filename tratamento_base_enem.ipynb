Codificando variáveis categóricas

```python
# Função gerada pelo GPT para codificar variáveis categóricas
import re

def encode_ordinal_last_unknown(
    df: pd.DataFrame,
    cols,
    *,
    suffix_ord: str = "_ord",
    suffix_flag: str = "_nao_sei",
    fillna_zero: bool = True,
    normalize: bool = True
) -> pd.DataFrame:
    """
    Transforma variáveis categóricas ordinais em (número, flag_nao_sei),
    assumindo que a ÚLTIMA letra presente (ex.: G ou H) corresponde à resposta "não sei".

    Regras:
      - Letras válidas: A–Z (caso-insensível, espaços são ignorados).
      - A..(última-1) viram 1..N (ordinais crescentes).
      - A última letra vira apenas a flag *_nao_sei = 1; o ordinal vira NaN (ou 0 se fillna_zero=True).
      - Valores fora de A–Z permanecem como NaN no ordinal e flag=0.

    Parâmetros:
      df           : DataFrame de entrada (modificado in-place).
      cols         : str ou lista de colunas a processar.
      suffix_ord   : sufixo para a coluna numérica ordinal.
      suffix_flag  : sufixo para a flag binária do "não sei".
      fillna_zero  : se True, preenche ordinais NaN com 0 (útil p/ regressão linear/RNA).
      normalize    : se True, faz strip + upper nas strings antes do processamento.

    Retorna:
      O próprio DataFrame (com as novas colunas adicionadas).
    """
    if isinstance(cols, str):
        cols = [cols]

    letter_regex = re.compile(r"^[A-Z]$")

    for col in cols:
        s = df[col].astype(object)

        # normaliza: tira espaços e sobe pra maiúscula
        if normalize:
            s = s.map(lambda x: x.strip().upper() if isinstance(x, str) else x)

        # letras únicas válidas presentes na coluna
        uniques = pd.Series(s.dropna().unique())
        letters = uniques[uniques.map(lambda x: isinstance(x, str) and bool(letter_regex.match(x)))]

        # se não houver letras válidas, cria colunas "vazias"
        if len(letters) == 0:
            df[col + suffix_ord]  = np.nan if not fillna_zero else 0
            df[col + suffix_flag] = 0
            continue

        # a última letra (lexicograficamente) é o "não sei"
        nao_sei_letter = max(letters)

        # mapeia A..(última-1) -> 1..N (mantém ordem)
        valid_letters = sorted([lt for lt in letters if lt != nao_sei_letter])
        mapping = {lt: i + 1 for i, lt in enumerate(valid_letters)}

        # aplica
        ord_col  = s.map(mapping)                      # última letra vira NaN aqui
        flag_col = (s == nao_sei_letter).astype(int)   # 1 se "não sei", 0 caso contrário

        # valores não mapeados (ex.: NaN original ou fora de A–Z) mantêm NaN no ordinal e 0 na flag
        flag_col = flag_col.fillna(0).astype(int)

        if fillna_zero:
            ord_col = ord_col.fillna(0).astype(int)

        df[col + suffix_ord]  = ord_col
        df[col + suffix_flag] = flag_col

    return df
```
